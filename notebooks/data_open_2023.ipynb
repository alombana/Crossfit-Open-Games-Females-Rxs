```python
import requests
import pandas as pd
import time
import functions
```

```python
open_2023='https://c3po.crossfit.com/api/leaderboards/v2/competitions/open/2023/leaderboards?view=0&division=2&region=0&scaled=0&sort=0&page=1'
response=requests.get(open_2023)
data=response.json()
```

```python
page=1
check=False
data_2023=[]
while check!=True:
    open_2023=f'https://c3po.crossfit.com/api/leaderboards/v2/competitions/open/2023/leaderboards?view=0&division=2&region=0&scaled=0&sort=0&page={page}'
    response=requests.get(open_2023)
    
    if (response.status_code==200):
        data=response.json()
        data_2023.append(data)
        
    else:
        raise ValueError('no access, status is not equal to 200')
    
    check=data['pagination']['totalPages']== data['pagination']['currentPage']
    page+=1
    time.sleep(0.2)

```

```python
final_data=functions.calling_file(data_2023)
```

```python
final_data # List with the information that will be used for creating the DataFrame for the Open games 2023
```

```python
rank=pd.DataFrame(final_data[0])
points=pd.DataFrame(final_data[1])
competitors_info=pd.DataFrame(final_data[2])
exerc1=pd.DataFrame(final_data[3])
exerc2=pd.DataFrame(final_data[4])
exerc3=pd.DataFrame(final_data[5])
exerc4=pd.DataFrame(final_data[6])
```

```python
df_open2023=pd.concat([rank, points,competitors_info,exerc1,exerc2,exerc3,exerc4], axis=1)
df_open2023['Year']='2023'
```
